```python
import pandas as pd
```

```python
#luetaan 2019 data
df = pd.read_csv('C:\\apu\\liikennemaara_h_2019-01-01-2019-12-31_viikinmaki.csv',sep =';',encoding= 'utf-8',header = 0)
df
```

```python
#dataframen muokkaus
```

```python
df = df.drop(['kaista','jaottelu','sijainti','ajoneuvoluokka'],axis=1)
```

```python
columns = df.columns
tila = columns[0:4]
tila2 = columns[4:-1]
```

```python
df_unpivot = pd.melt(df, id_vars=tila, value_vars=tila2)
```

```python
df_unpivot.rename(columns = {'value':'Määrä'},inplace=True)
df_unpivot.rename(columns = {'variable':'klo'},inplace=True)
```

```python
df_unpivot[['klo', 'poisto']] = df_unpivot['klo'].str.split('_', n=1, expand=True)
df = df_unpivot.drop(['poisto'],axis =1)
```

```python
#lisätään sarake keskinopeus ja muutetaan järjestystä
df['keskinopeus'] = 0
df.rename(columns = {'pistetunnus':'sijainti'},inplace=True)
df19 = df[['sijainti','pvm', 'klo','suunta','suuntaselite', 'Määrä','keskinopeus']]
df19
```

```python
#luetaan 2020-23 data
df4 = pd.read_csv('C:\\apu\\keskinopeus_h_2020-01-01-2023-03-31_viikinmaki.csv', sep=';')
df4
```

```python
#Muokataan dataframea
df20_23 = df4.drop(['sijainti','nopeusrajoitus','kaikki', 'raskaat','kaikki','keskinopeus_kaikki','keskinopeus_raskaat'], axis=1)
df20_23.rename(columns = {'pistetunnus':'sijainti'},inplace=True)
df20_23.rename(columns = {'kevyet':'Määrä'},inplace=True)
df20_23.rename(columns = {'keskinopeus_kevyet':'keskinopeus'},inplace=True)
df20_23
```

```python
#Yhdistetään 2019 ja 20_23 
dfLiikenne = pd.concat([df19, df20_23], ignore_index=True)
dfLiikenne
```

```python
#tallennus csv -tiedostoon
dfLiikenne.to_csv('Liikennemaarat_nopeudet_2019_23.csv')
```
