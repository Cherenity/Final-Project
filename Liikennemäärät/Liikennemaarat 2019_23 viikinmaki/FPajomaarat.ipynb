```python
import pandas as pd
```

```python
df = pd.read_csv('C:/FinalProject/liikennemaara_h_2019-01-01-2019-12-31.csv',sep =';',encoding= 'utf-8',header = 0)
df
```

```python
df = df.drop(['kaista','jaottelu','pistetunnus','ajoneuvoluokka'],axis=1)
df
```

```python
columns = df.columns
tila = columns[0:4]
tila2 = columns[4:-1]
```

```python
df_unpivot = pd.melt(df, id_vars=tila, value_vars=tila2)
```

```python
df_unpivot.rename(columns = {'value':'Määrä'},inplace=True)
df_unpivot.rename(columns = {'variable':'Klo'},inplace=True)
```

```python
df_unpivot[['Klo', 'poisto']] = df_unpivot['Klo'].str.split('_', n=1, expand=True)
df = df_unpivot.drop(['poisto'],axis =1)
```
