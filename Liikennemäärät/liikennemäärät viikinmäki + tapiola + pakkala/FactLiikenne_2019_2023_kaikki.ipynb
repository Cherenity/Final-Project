```python
import pandas as pd
```

```python
#luetaan 2019 data
df = pd.read_csv('C:\\apu\\liikennemaara_h_2019-01-01-2019-12-31_viikinmaki.csv',sep =';',encoding= 'utf-8',header = 0)
df
```

```python
dftapiola19 = pd.read_csv('C:\\apu\\liikennemaara_h_2019-01-01-2019-12-31_tapiola.csv',sep =';',encoding= 'utf-8',header = 0)
dftapiola19
```

```python
dfpakkala19 = pd.read_csv('C:\\apu\\liikennemaara_h_2019-01-01-2019-12-31_pakkala.csv',sep =';',encoding= 'utf-8',header = 0)
dfpakkala19
```

```python
#yhdistetään kaikki 2019 df:t
df = pd.concat([df, dftapiola19, dfpakkala19])
df
```

```python
df = df.drop(['kaista','jaottelu','sijainti','ajoneuvoluokka'],axis=1)
```

```python
columns = df.columns
tila = columns[0:4]
tila2 = columns[4:-1]
```

```python
df_unpivot = pd.melt(df, id_vars=tila, value_vars=tila2)
```

```python
df_unpivot.rename(columns = {'value':'Määrä'},inplace=True)
df_unpivot.rename(columns = {'variable':'klo'},inplace=True)
```

```python
df_unpivot[['klo', 'poisto']] = df_unpivot['klo'].str.split('_', n=1, expand=True)
df = df_unpivot.drop(['poisto'],axis =1)
```

```python
#lisätään keskinopeus sarake ja muutetaan järjestystä
df['keskinopeus'] = 0
df.rename(columns = {'pistetunnus':'sijainti'},inplace=True)
df19 = df[['sijainti','pvm', 'klo','suunta','suuntaselite', 'Määrä','keskinopeus']]
df19
```

```python
#luetaan 2020-22 data
df4 = pd.read_csv('C:\\apu\\keskinopeus_h_2020-01-01-2023-03-31_viikinmaki.csv', sep=';')
```

```python
df4
```

```python
#Muokataan dataframea
df20_22 = df4.drop(['sijainti','nopeusrajoitus','kaikki', 'raskaat','kaikki','keskinopeus_kaikki','keskinopeus_raskaat'], axis=1)
df20_22.rename(columns = {'pistetunnus':'sijainti'},inplace=True)
df20_22.rename(columns = {'kevyet':'Määrä'},inplace=True)
df20_22.rename(columns = {'keskinopeus_kevyet':'keskinopeus'},inplace=True)
df20_22
```

```python
#luetaan tapiolan data 20-23
dftapiola20_23 = pd.read_csv('C:\\apu\\keskinopeus_h_2020-01-01-2023-03-31_tapiola.csv', sep=';')
dftapiola20_23
```

```python
#luetaan pakkalan data 20-23
dfpakkala20_23 = pd.read_csv('C:\\apu\\keskinopeus_h_2020-01-01-2023-03-31_pakkala.csv', sep=';')
dfpakkala20_23
```

```python
#yhdistetään tapiola ja pakkala 20-23
dftp20_23 = pd.concat([dftapiola20_23, dfpakkala20_23], ignore_index=True)
dftp20_23
```

```python
#Muokataan tapiola&pakkala dataframea
dftp20_23 = dftp20_23.drop(['sijainti','nopeusrajoitus','kaikki', 'raskaat','keskinopeus_kaikki','keskinopeus_raskaat'], axis=1)
dftp20_23.rename(columns = {'pistetunnus':'sijainti'},inplace=True)
dftp20_23.rename(columns = {'kevyet':'Määrä'},inplace=True)
dftp20_23.rename(columns = {'keskinopeus_kevyet':'keskinopeus'},inplace=True)
dftp20_23
```

```python
# poistetaan null rivit
dftp20_23.dropna(inplace=True, axis=0)

# rivien määrän null poiston jälkeen
row_count_tp = len(dftp20_23)
print(row_count_tp)

dftp20_23.head(1)
```

```python
#muutetaan Määrä sarakkeen datatyyppi float to int
dftp20_23['Määrä'] = dftp20_23['Määrä'].astype(int)
dftp20_23.dtypes
```

```python
#Yhdistetään 2019 ja kumpula 20_22 ja tapiola&pakkala 20_23
dfLiikenne = pd.concat([df19, df20_22, dftp20_23], ignore_index=True)
dfLiikenne
```

```python
#csv:ksi
dfLiikenne.to_csv('c:\\apu\\Liikennemaarat_nopeudet_2019_23.csv')
```
